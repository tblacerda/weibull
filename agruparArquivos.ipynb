```python
# Importando os módulos necessários
import pandas as pd
import os

# Definindo o diretório onde estão as planilhas excel
diretorio = "./dadosDiarios/"

# Criando uma lista vazia para armazenar os dados de cada planilha
dados = []

# Percorrendo os arquivos do diretório
for arquivo in os.listdir(diretorio):
    # Verificando se o arquivo é uma planilha excel
    if arquivo.endswith(".xlsx"):
        # Lendo a planilha e adicionando à lista de dados
        planilha = pd.read_excel(os.path.join(diretorio, arquivo))
        dados.append(planilha)

# Concatenando os dados de todas as planilhas em um único dataframe
df = pd.concat(dados, ignore_index=True)

# Eliminando as linhas duplicadas do dataframe
df = df.drop_duplicates()
df = df[['RAN Node','Dia', 'DISP_COUNTER_TOTAL 4G (com filtro OPER)']]
df.columns = ['Site', 'Date', 'Disponibilidade']
df['Disponibilidade'] = df['Disponibilidade'].fillna(0)
```

```python
df.to_excel('dadosDisponibilidade.xlsx', index=False)
```
